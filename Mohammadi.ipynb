# 39. "pd.pivot_table()" metodunu kullanarak, dataFrame'deki bir sütundan tek indeksli "pivot table" oluşturunuz.


```python
print("\n--- 39. Adım: Tek İndeksli Pivot Table ---")
# Dinamik olması için; indeks olarak ilk kategorik (metin) sütunu, hesaplama (values) için ilk sayısal sütunu seçiyoruz.
kat_sutun_1 = df_Mohammadi.select_dtypes(exclude=['number']).columns[0]
num_sutun_1 = df_Mohammadi.select_dtypes(include=['number']).columns[0]

# aggfunc='mean' ile o kategorilerdeki ortalama sayısal değerleri alıyoruz.
pivot_tek = pd.pivot_table(df_Mohammadi, values=num_sutun_1, index=kat_sutun_1, aggfunc='mean')
print(f"'{kat_sutun_1}' bazında '{num_sutun_1}' ortalamaları:")
display(pivot_tek.head())
```

# 40. "pd.pivot_table()" metodunu kullanarak, dataFrame'deki birden fazla sütundan çok indeksli "pivot table" oluşturunuz.


```python
print("\n--- 40. Adım: Çok İndeksli Pivot Table ---")
# İkinci bir kategorik sütun daha seçelim. (Veri setinde yeterince kategorik sütun olduğu varsayımıyla)
kat_sutun_2 = df_Mohammadi.select_dtypes(exclude=['number']).columns[1]

# Liste içine birden fazla sütun adı vererek hiyerarşik (multi-index) bir tablo oluşturuyoruz.
pivot_coklu = pd.pivot_table(df_Mohammadi, values=num_sutun_1, index=[kat_sutun_1, kat_sutun_2], aggfunc='mean')
print(f"'{kat_sutun_1}' ve '{kat_sutun_2}' bazında '{num_sutun_1}' ortalamaları:")
display(pivot_coklu.head(10))
```

#71. 72. 73. Data Wrangling, Cleaning & Cleansing
 Bu üç terim teknik olarak birbirinin yerine kullanılsa da genel yaklaşım şöyledir:
 Wrangling: Ham veriyi yapılandırılmış hale getirmek.
 Cleaning: Hatalı, eksik veya tutarsız verileri düzeltmek.
 Cleansing: Veriyi standartlara uygun (kural tabanlı) hale getirmek.

```python
import pandas as pd
import duckdb
import sqlite3

print("\n--- 71-73: Data Wrangling & Cleaning (Genel İşlemler) ---")
# 1. Wrangling: Sütun adlarını düzelt ve boşlukları yönet
df_wrangled = df_Mohammadi.copy()
df_wrangled.columns = [c.lower().replace(' ', '_') for c in df_wrangled.columns]

# 2. Cleaning: Eksik verileri doldurma (Sayısal için medyan, kategorik için 'Bilinmiyor')
for col in df_wrangled.select_dtypes(include=['number']).columns:
    df_wrangled[col] = df_wrangled[col].fillna(df_wrangled[col].median())
for col in df_wrangled.select_dtypes(include=['object']).columns:
    df_wrangled[col] = df_wrangled[col].fillna('Bilinmiyor')

# 3. Cleansing: Veri formatını standartlaştırma (Örn: 'state' sütununu büyük harfe çevir)
if 'state' in df_wrangled.columns:
    df_wrangled['state'] = df_wrangled['state'].str.upper()

print("Veri temizleme ve düzenleme tamamlandı.")
display(df_wrangled.head(3))
```

#74. DuckDB Kullanımı


```python
print("\n--- 74: DuckDB Veritabanı ---")
# DuckDB, analitik sorgular için cực kỳ (aşırı) hızlı, dosya tabanlı bir veritabanıdır.
con = duckdb.connect(database='banka_verisi.duckdb', read_only=False)

# DataFrame'i DuckDB'ye tablo olarak yükle
con.execute("CREATE OR REPLACE TABLE hesaplar AS SELECT * FROM df_wrangled")

# DuckDB üzerinde SQL sorgusu çalıştır
duck_sonuc = con.execute("SELECT account_type, AVG(age) FROM hesaplar GROUP BY account_type").fetchdf()
print("DuckDB üzerinden alınan sonuç:")
display(duck_sonuc)
con.close()
```

#75. SQLite Kullanımı


```python
print("\n--- 75: SQLite Veritabanı ---")
# SQLite standart veritabanı yönetimi için harikadır.
conn = sqlite3.connect('banka_verisi.sqlite')

# DataFrame'i SQLite tablosuna yaz
df_wrangled.to_sql('hesaplar_sqlite', conn, if_exists='replace', index=False)

# SQL sorgusu çalıştır
sqlite_sonuc = pd.read_sql("SELECT state, COUNT(*) FROM hesaplar_sqlite GROUP BY state LIMIT 5", conn)
print("SQLite üzerinden alınan sonuç:")
display(sqlite_sonuc)
conn.close()
```
